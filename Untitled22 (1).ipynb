```python
from pymongo import MongoClient

client = MongoClient("mongodb+srv://kathircts673_db_user:CjyBddeHgoYCNhW8@cluster0.tcpigsu.mongodb.net/")# Replace with your MongoDB connection string
db = client['kaiser_db']
dataset_queries = db['dataset_queries']
coverage_results = db['coverage_results']

dataset_id = "D001"

# Step 1: Fetch all records for this dataset
records = list(dataset_queries.find({"dataset_id": dataset_id}))

# Step 2: Group counts by topic, sub-topic, and section
from collections import defaultdict

counts = defaultdict(int)

for r in records:
    key = (r["topic"], r["sub_topic"], r["section"])
    counts[key] += 1

counts
```

```python
# Expected queries (dummy values for now)
expected_counts = {
    ('Cardiology Care Guide', 'CHEST PAIN, ACUTE AND STABLE', 'Evaluation'): 5,  # Replace with real value from Vamsi
}

# Step 3: Calculate coverage and insert into the coverage_results collection
for (topic, sub_topic, section), actual_count in counts.items():
    expected_count = expected_counts.get((topic, sub_topic, section), 0)

    # Calculate coverage percentage
    coverage_percentage = (actual_count / expected_count) * 100 if expected_count > 0 else 0

    # Insert coverage result into MongoDB
    coverage_results.insert_one({
        "dataset_id": dataset_id,
        "topic": topic,
        "sub_topic": sub_topic,
        "section": section,
        "actual": actual_count,
        "expected": expected_count,
        "coverage_percentage": coverage_percentage
    })

print("Coverage results inserted successfully!")
```

```python
import pandas as pd

# Retrieve data from MongoDB collections
dataset_queries_data = list(dataset_queries.find({"dataset_id": dataset_id}))
coverage_results_data = list(coverage_results.find({"dataset_id": dataset_id}))

# Convert dataset_queries to DataFrame
dataset_queries_df = pd.DataFrame(dataset_queries_data)

# Convert coverage_results to DataFrame
coverage_results_df = pd.DataFrame(coverage_results_data)

# Export DataFrames to Excel
with pd.ExcelWriter("/content/output_data.xlsx") as writer:
    dataset_queries_df.to_excel(writer, sheet_name="Dataset Queries", index=False)
    coverage_results_df.to_excel(writer, sheet_name="Coverage Results", index=False)

print("Data exported to Excel successfully!")
```
